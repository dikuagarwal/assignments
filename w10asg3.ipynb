Q1: What is Estimation Statistics? Explain point estimate and interval estimate.<br>
Ans. Estimation statistics is a branch of inferential statistics that focuses on making inferences about population parameters based on sample data. A value or a range of value is used to esimate the unknown population parameter based on sample parameter. Normally there are two types of estimate: point and interval estimate. In point estimate a single value is represent the estimated value of unknown population parameter. An interval estimate, also known as a confidence interval, is a range of values within which the unknown population parameter is likely to fall, with a certain level of confidence. Confidence intervals are often associated with a confidence level, typically expressed as a percentage (e.g., 95% confidence interval).<hr>

Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.<br>
Ans. According to CLT, the mean of the sampling distribution of x̄ is equal to the population mean, μ, and that the standard deviation is given by σ/√n, where σ is the population standard deviation. The standard deviation of a sampling distribution is called the standard error.
In the large-sample case, at a 95% confidence interval estimate for the population mean is given by x̄ ± 1.96σ/√n. When the population standard deviation, σ, is unknown, the sample standard deviation is used to estimate σ in the confidence interval formula. The quantity 1.96σ/√n is often called the margin of error for the estimate.

```python
sample_mean = 75.2
samle_std = 5.8
sample_size = 100
import math
def pm(m,std,s):
    standard_error = std/ math.sqrt(s)
    margin_of_error = 1.96 * standard_error  # 95% confidence interval
    upper_bound = m + margin_of_error
    lower_boound = m - margin_of_error
    
    print(f"Estimated Population mean: {m}")
    print(f"At a 95% confidence interval, Estimate mean is given by [{lower_boound}, {upper_bound}]")
    
pm(sample_mean,samle_std,sample_size)
```

given a sample from each population.<br>

```python
import numpy as np
from scipy.stats import ttest_ind

# Example usage:
null_hypo = " There is no significant difference between the population means."
sample1 = [10, 15, 20, 25, 30]  # Sample data from the first population
sample2 = [12, 18, 22, 28, 32]  # Sample data from the second population
alpha = 0.05

tstat, pvalue = ttest_ind(sample1,sample2,equal_var=True)
if pvalue < alpha:
    print("reject the null hypothesis")
else: 
    print(f"{pvalue}, we fail to reject null hypothesis")
```

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Set the degrees of freedom
df = 10

# Generate x values (t-values) for the plot
x = np.linspace(-4, 4, 500)  # Choose a range that covers the tails of the t-distribution

# Calculate the probability density function (PDF) using the t-distribution
pdf = t.pdf(x, df)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, pdf, label='t-distribution (df=10)')
plt.xlabel('t-values')
plt.ylabel('Probability Density Function (PDF)')
plt.title("Student's t-distribution with df=10")
plt.grid()
plt.legend()
plt.show()
```

- Whether the two samples data groups are independent.

<hr>Q11: What is Student’s t distribution? When to use the t-Distribution.<br>
Ans. Student T distribution or simple t distribution is a probality distribution that is widely used in hypothesis testing and confidence interval estimation when dealing with small sample size or when the population std is unknown. The t-distribution is bell-shaped like the standard normal (Z) distribution, but it has thicker tails.<br>
The t-distribution is used in situations where:<br>
- Sample Size is Small: The t-distribution is preferred when dealing with small sample sizes (typically n < 30). For larger sample sizes, the t-distribution approximates the standard normal distribution, and using Z-scores is generally acceptable.
- Population Standard Deviation is Unknown: When the population standard deviation (σ) is unknown, and you have to estimate it using the sample standard deviation (s).
- Hypothesis Testing and Confidence Intervals: The t-distribution is commonly used in hypothesis testing and constructing confidence intervals for the population mean when the sample size is small and the population standard deviation is unknown.

Q12: What is t-statistic? State the formula for t-statistic.<br>
Ans. In a one-sample t-test, the t-statistic measures how many standard errors the sample mean is away from the hypothesized population mean. In a two-sample t-test, the t-statistic compares the difference between the population means of the two sample groups.<br>
The formula for the t-statistic depends on the type of test being performed, but for a one-sample t-test or a two-sample t-test assuming equal variances, the formula is as follows:

For a one-sample t-test:
t = (x̄ - μ) / (s / √n)

For a two-sample t-test (assuming equal variances):
t = (x̄1 - x̄2) / √((s1^2 / n1) + (s2^2 / n2))<hr>

<hr>Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.<br>
Ans. Step 1: Set up hypotheses:<br>
Null hypothesis: a new drug will decrease blood pressure by 10 mmHg.<br>
ALternative Hypothesis: a new drug will not decrease blood pressure by 10 mmHg or (μ) ≠ 10 mmHg . (2 tail test).<br>
Step 2: Set the significance level (α): alpha or significant value is 0.05<br>
step3: Select the appropiate Statistic test: We will perfrom one-sample t test because of we don't have population std.<br>


```python
# step4: Calculate the t or test statistics:
sample_mean = 8
sample_size = 100
sample_std = 3
u = 10  # Hypothesized_population_mean
# The formula for the t-statistic in a one-sample t-test is: t = (x̄ - μ) / (s / √n)
import math 
ts = abs( (sample_mean - u) / (sample_std / math.sqrt(sample_size))  )
print(ts)

# Step5: Determine critical value: 
alpha = 0.05
df = sample_size - 1

from scipy.stats import t

critical_t = abs (t.ppf(1 - alpha / 2, df)  ) # (for one-tailed tests, use (alpha ) instead of (1- alpha/2)  # take absolue value
print(critical_t)

#Step6:  Compare the test statistic with the critical value:
flag = 0
if critical_t < ts:
    print("This means that the test statistic falls into the critical region, and we reject the null hypothesis.")
    flag = 1
else:
    print("We fail to reject the Null hypothesis")

# step7: Make a conclusion
if flag==1:
    print("the new drug has a significant effect on reducing blood pressure by an amount different from the hypothesized 10 mmHg.")
else:
    print("the new drug has a significant effect on reducing blood pressure by an amount  10 mmHg.")
```

<hr>Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.<br>
Ans. Null Hypo: Product mean weight >= 5<br>
Alternative hypo: Product mean weight < 5 (1 tail test)<br>
Significant value = 0.01<br>
We will perfrom Z test because we have population std and sample size is small.<br>

```python
import numpy as np
from scipy.stats import norm

# Given data
population_mean = 5
population_std = 0.5
sample_size = 25
sample_mean = 4.8

# Calculate the standard error of the mean (SEM)
SEM = population_std / np.sqrt(sample_size)

# Calculate the Z-score
Z_score = (sample_mean - population_mean) / SEM
print(Z_score)

# Set the significance level (alpha)
alpha = 0.01

# Find the critical Z-value for a one-tailed test
critical_Z = norm.ppf(alpha)   # or 1 -alpha and for 2-tail test alpha/2
print(critical_Z)
# Make a decision
if Z_score < critical_Z:
    print("Reject the null hypothesis. The true mean weight of the products is less than 5 pounds.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the true mean weight is less than 5 pounds.")
```
