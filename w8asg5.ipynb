Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.<br>
Ans. In Bokeh, glyphs are visual representations of data on a plot. They are used to display various shapes, such as lines, markers, patches, or images, to represent data points or patterns in a plot. Glyphs allow you to customize the appearance and properties of your data on a Bokeh plot.

To add glyphs to a Bokeh plot, you need to use the Glyph API provided by Bokeh. The Glyph API provides a wide range of glyph types, each with its own set of properties that can be customized.

Here's an example of how to add glyphs to a Bokeh plot using the circle() glyph:

```python
from bokeh.plotting import figure, show

# Create a new figure
p = figure(min_width=400, min_height=300)

# Define some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Add a circle glyph to the plot
p.circle(x, y, size=10, color="navy", alpha=0.5)   # we add a circle glyph to the plot using the circle() method

# Show the plot
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?<br>
Ans. A Bokeh server is a Python server application that allows you to create interactive Bokeh plots that can be updated in real time. It enables you to build dynamic web applications with interactive visualizations by combining Bokeh plots with other Python functionality.<br>
To use the Bokeh server and create interactive plots, you need to follow these steps:

- Define your plot and its components: Start by creating your Bokeh plot using the Bokeh plotting API, just as you would for a static plot.

- Define your application code: Next, you need to define the application code that will handle the interactions and updates for your plot. This code typically resides in a Python script file. It involves creating a function that defines the necessary callbacks and updates to be performed when user interactions occur.

- Run the Bokeh server: Once you have your plot and application code ready, you can run the Bokeh server to serve your interactive plot. You can start the server by running the Python script file that contains your application code.

```python
from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from random import random

# Create a figure and data source
p = figure()
source = ColumnDataSource(data=dict(x=[0], y=[0]))

# Create a circle glyph using the data source
p.circle('x', 'y', size=10, color="navy", alpha=0.5, source=source)

# Define the callback function to update the plot
def update():
    new_data = dict(x=[random()], y=[random()])
    source.data = new_data

# Add a periodic callback to update the plot every 1 second
curdoc().add_periodic_callback(update, 1000)

# Run the Bokeh server
curdoc().title = "Interactive Plot"
curdoc().add_root(p)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?<br>
Ans. To embed a Bokeh plot into a web page or dashboard using Flask or Django, you can follow these general steps:

- Create the Bokeh plot: Start by creating your Bokeh plot using the Bokeh plotting API, just as you would for a standalone plot.

- Convert the Bokeh plot to HTML: Use the bokeh.embed module to convert the Bokeh plot to an HTML document that can be embedded in a web page.

- Set up your web framework: Depending on whether you are using Flask or Django, you need to set up your web framework to serve the web page or dashboard where the Bokeh plot will be embedded.

- Pass the Bokeh plot to the web page: In your web framework, you need to pass the HTML document of the Bokeh plot to the template or view that will render the web page. This can be done by injecting the HTML code into the appropriate context variables.

- Render the web page: Finally, render the web page or dashboard template in your web framework, making sure to include the context variables that contain the Bokeh plot's HTML.
