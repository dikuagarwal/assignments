Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.<br>
Ans. scatter plot is one kind of plot that we can draw with help of matplotlib libraray. it's a graph in which the values of two variables are plotted along two axes, the pattern of the resulting points revealing any correlation present.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
x,y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:<br>
Ans. The subplots() function in the Matplotlib acts as a utility wrapper. This function helps in creating common layouts of subplots and it also includes the enclosing figure object, in a single call.

```python
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)   # (row,coumn,nth plot)
plt.plot(x,y)
plt.title("first plot")

y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Second plot")

y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Third plot")

y = y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Fourth plot")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.<br>
Ans. A bar plot, on the other hand, is used to represent categorical data, which is data that can be divided into distinct categories. In a bar plot, each category is represented by a separate bar, and the height of the bar represents the frequency or count of data points in that category.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

Q5: What is a box plot? Why is it used?<br>
Ans. A Box and Whisker Plot (or Box Plot) is a convenient way of visually displaying the data distribution through their quartiles. It is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
```
